```python
import numpy as np
import pandas as pd

# set random seed for reproducibility
np.random.seed(123)

# create a DataFrame with random data of two classes
n_samples = 1000
class_0_ratio= 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```
